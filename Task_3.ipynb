```python

# 📊 TASK 3: Data Visualization (Colab Ready)


# 1️⃣ Install & Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# 2️⃣ Load a Sample Dataset (Titanic Dataset from seaborn)
df = sns.load_dataset("titanic")
print("Dataset Loaded ✅")
print(df.head())
print("\nBasic Info:")
print(df.info())

# 3️⃣ Handle Missing Data
df = df.dropna(subset=["age", "fare", "sex", "class", "survived"])

# 4️⃣ Visualization 1 - Countplot (Survival Count)
plt.figure(figsize=(6,4))
sns.countplot(x="survived", data=df, palette="pastel")
plt.title("Survival Count")
plt.savefig("survival_count.png")
plt.show()

# 5️⃣ Visualization 2 - Histogram (Age Distribution)
plt.figure(figsize=(6,4))
sns.histplot(df["age"], kde=True, bins=20, color="skyblue")
plt.title("Age Distribution")
plt.savefig("age_distribution.png")
plt.show()

# 6️⃣ Visualization 3 - Boxplot (Fare by Class)
plt.figure(figsize=(6,4))
sns.boxplot(x="class", y="fare", data=df, palette="Set3")
plt.title("Fare by Passenger Class")
plt.savefig("fare_by_class.png")
plt.show()

# 7️⃣ Visualization 4 - Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df[["age","fare","survived"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig("correlation_heatmap.png")
plt.show()

# 8️⃣ Visualization 5 - Pairplot (Age, Fare, Survived)
sns.pairplot(df[["age","fare","survived"]], hue="survived", palette="husl")
plt.savefig("pairplot.png")
plt.show()

# 9️⃣ Generate Summary File
summary_text = f"""
TASK 3 - DATA VISUALIZATION SUMMARY
-----------------------------------
Dataset Used: Titanic (Seaborn)

Insights:
1. Survival Count: More passengers did NOT survive than survived.
2. Age Distribution: Most passengers were between 20–40 years old.
3. Fare by Class: First-class passengers paid much higher fares.
4. Correlation: Fare and survival have a positive correlation.
5. Pairplot: Shows separation between high-fare passengers (more likely survived).

Total Records Used: {df.shape[0]}
"""
with open("visualization_summary.txt", "w") as f:
    f.write(summary_text)

print("✅ All visualizations saved and summary created!")

# 🔽 Download Files (You can comment these lines if running locally)
files.download("survival_count.png")
files.download("age_distribution.png")
files.download("fare_by_class.png")
files.download("correlation_heatmap.png")
files.download("pairplot.png")
files.download("visualization_summary.txt")
```
